```python
# Three-Hump Camel Function

class ThreeHumpCamel:
  '''
  Three Hump Camel
 
  Adapted from https://github.com/keit0222/optimization-evaluation/blob/master/opteval/benchmark_func.py
  '''
  x_domain = np.linspace(-5, 5) 
  y_domain = np.linspace(-5, 5)
  global_minimum = np.array([[0, 0]])

  @staticmethod    
  def fun(x):
    return 2*np.power(x[0],2) - 1.05*np.power(x[0],4) + np.power(x[0],6)/6 + x[0]*x[1] + np.power(x[1],2)
```

```python
#Six-Hump Camel Function 
class SixHumpCamel:
  '''
  Six-Hump Camel Function
 
  Adapted from https://github.com/keit0222/optimization-evaluation/blob/master/opteval/benchmark_func.py
  '''
  x_domain = np.linspace(-3, 3)
  y_domain = np.linspace(-2, 2)
  global_minimum = np.array([[.0898, -.7126],
                             [-.0898, .7126]])

  @staticmethod
  def fun(x):
      return (4-2.1*np.power(x[0],2)+(np.power(x[0],4)/3))*np.power(x[0],2) + (x[0]*x[1]) + (4*np.power(x[1],2)-4)*np.power(x[1],2)
```

```python
# Dixon-Price Function


class DixonPrince:
  '''
    Adapted From https://gist.github.com/denis-bz/da697d8bc74fae4598bf
  '''
  
  dimension = [np.linspace(-10, 10) for i in range(10)]
  domain = []
  
  @staticmethod
  def fun(x):
    x = np.asarray_chkfinite(x)
    n = len(x)
    j = np.arange( 2, n+1 )
    x2 = 2 * x**2
    return sum( j * (x2[1:] - x[:-1]) **2 ) + (x[0] - 1) **2
```

```python
from itertools import product

def gen_guesses(x_domains, n=10):
  '''
  >>> x = np.array([-5, 5])
  >>> y = np.array([5, -5])
  >>> gen_guesses((x, y), 2)
  array([[-5.,  5.],
         [-5.,  0.],
         [-5., -5.],
         [ 0.,  5.],
         [ 0.,  0.],
         [ 0., -5.],
         [ 5.,  5.],
         [ 5.,  0.],
         [ 5., -5.]])
  '''
  guesses = np.zeros((len(x_domains), n))
  for i in range(len(x_domains)):
    guesses[i, :] = np.linspace(x_domains[i][0], x_domains[i][-1], num=n)

  return np.array(list(product(*guesses)))
```

## Three Hump Camel Experiments

```python
# Three-Hump Camel Function 3D Plot

fig = plt.figure(figsize=(21, 9), facecolor='w')
fig.suptitle('Three-Hump Camel Function 3D Plot', size=20)
fig.subplots_adjust(wspace=0)

x = np.linspace(-5, 5)
y = np.linspace(-5, 5)

X, Y = np.meshgrid(x, y)
Z = ThreeHumpCamel.fun((X, Y))

ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(X, Y, Z, cmap='jet')
ax.view_init(azim=-120)
ax.set_xlabel('x0')
ax.set_ylabel('x1')
ax.set_zlabel('f(x0, x1)', labelpad=10)

x = np.linspace(-2, 2)
y = np.linspace(-2, 2)

X, Y = np.meshgrid(x, y)
Z = ThreeHumpCamel.fun((X, Y))

ax = fig.add_subplot(122, projection='3d')

ax.plot_surface(X, Y, Z, cmap='jet')
ax.view_init(azim=-120)
ax.set_xlabel('x0')
ax.set_ylabel('x1')
ax.set_zlabel('f(x0, x1)')

plt.show()
```

```python
from scipy.optimize import minimize
from numdifftools import Jacobian, Hessian

fun = ThreeHumpCamel.fun

def fun_der(x):
  '''fun's derivative'''
  return Jacobian(lambda x: fun(x))(x).ravel()

def fun_hess(x):
  '''fun's 2nd derivative'''
  return Hessian(lambda x: fun(x))(x)
```

### Performance (Time and number of execution)

```python
# Steepest Descent Performance (Time and number of execution)

guess = np.array([-5. , -1])
%timeit minimize(ThreeHumpCamel.fun, guess, jac=fun_der, method='CG', tol=1e-16)
minimize(ThreeHumpCamel.fun, guess, jac=fun_der, method='CG', tol=1e-16)
```

```python
# Quasi Newton Performance (Time and number of execution)

guess = np.array([.5, 1.6])
%timeit minimize(ThreeHumpCamel.fun, guess, jac=fun_der, method='Newton-CG', tol=1e-16)
minimize(ThreeHumpCamel.fun, guess, jac=fun_der, method='Newton-CG', tol=1e-16)
```

```python
# Newton Performance (Time and number of execution)

guess = np.array([.5, 1.6])
%timeit minimize(ThreeHumpCamel.fun, guess, jac=fun_der, hess=fun_hess, method='Newton-CG', tol=1e-16)
minimize(ThreeHumpCamel.fun, guess, jac=fun_der, hess=fun_hess, method='Newton-CG', tol=1e-16)
```

### Guess Initialization

```python
# Steepest Descent Guess Initialization

for guess in gen_guesses((ThreeHumpCamel.x_domain, ThreeHumpCamel.y_domain)):
  res = minimize(ThreeHumpCamel.fun, guess, jac=fun_der, method='CG', tol=1e-16)
  print(res.success, guess, res.x, res.fun)
```

```python
# Quasi Newton Guess Initialization

for guess in gen_guesses((ThreeHumpCamel.x_domain, ThreeHumpCamel.y_domain)):
  res = minimize(ThreeHumpCamel.fun, guess, jac=fun_der, method='Newton-CG', tol=1e-16)
  print(res.success, guess, res.x, res.fun)
```

```python
# Newton Guess Initialization

for guess in gen_guesses((ThreeHumpCamel.x_domain, ThreeHumpCamel.y_domain)):
  res = minimize(ThreeHumpCamel.fun, guess, jac=fun_der, hess=fun_hess, method='Newton-CG', tol=1e-16)
  print(res.success, guess, res.x, res.fun)
```

## Six Hump Camel Experiment

```python
# Six-Hump Camel Function 3D Plot

fig = plt.figure(figsize=(21, 9), facecolor='w')
fig.suptitle('Six-Hump Camel Function 3D Plot', size=20)
fig.subplots_adjust(wspace=0)

x = np.linspace(-3, 3)
y = np.linspace(-2, 2)

X, Y = np.meshgrid(x, y)
Z = SixHumpCamel.fun((X, Y))

ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(X, Y, Z, cmap='jet')
ax.view_init(azim=-120)
ax.set_xlabel('x0')
ax.set_ylabel('x1')
ax.set_zlabel('f(x0, x1)', labelpad=10)

x = np.linspace(-2, 2)
y = np.linspace(-1, 1)

X, Y = np.meshgrid(x, y)
Z = SixHumpCamel.fun((X, Y))

ax = fig.add_subplot(122, projection='3d')

ax.plot_surface(X, Y, Z, cmap='jet')
ax.view_init(azim=-130, elev=30)
ax.set_xlabel('x0')
ax.set_ylabel('x1')
ax.set_zlabel('f(x0, x1)')
ax.set_zlim([-2,6])

plt.show()
```

```python
fun = SixHumpCamel.fun

def fun_der(x):
  '''fun's derivative'''
  return Jacobian(lambda x: fun(x))(x).ravel()

def fun_hess(x):
  '''fun's 2nd derivative'''
  return Hessian(lambda x: fun(x))(x)

x0 = np.array([0.0898, -0.7126])
x1 = np.array([-0.0898, 0.7126])
guess = gen_guesses((x0, x1), 1)
res = minimize(fun, guess, jac=fun_der, options={'disp': True}, tol=1e-16)
print(guess)
print(res)
```

```python
# Newton Guess Initializations

for guess in gen_guesses((SixHumpCamel.x_domain, SixHumpCamel.y_domain)):
  res = minimize(SixHumpCamel.fun, guess, jac=fun_der, hess=fun_hess, method='Newton-CG', tol=1e-32)
  print(res.success, guess, res.x, res.fun)
```

```python
# Quasi Newton Guess Initializations

for guess in gen_guesses((SixHumpCamel.x_domain, SixHumpCamel.y_domain)):
  res = minimize(SixHumpCamel.fun, guess, jac=fun_der, method='Newton-CG', tol=1e-10)
  print(res.success, guess, res.x, res.fun)
```

```python
print('Initial guess:', SixHumpCamel.global_minimum[0])
res = minimize(SixHumpCamel.fun, SixHumpCamel.global_minimum[0], jac=fun_der, hess=fun_hess, method='Newton-CG')
print(res)
print('Initial guess:', SixHumpCamel.global_minimum[1])
res = minimize(SixHumpCamel.fun, SixHumpCamel.global_minimum[1], jac=fun_der, hess=fun_hess, method='Newton-CG')
print(res)
print(SixHumpCamel.fun(SixHumpCamel.global_minimum[0]))
print(SixHumpCamel.fun(SixHumpCamel.global_minimum[1]))
```
